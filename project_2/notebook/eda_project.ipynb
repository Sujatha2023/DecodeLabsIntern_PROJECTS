```python
plt.figure(figsize=(8,5))

sns.countplot(x="PaymentMethod", data=df)

plt.title("Payment Method Distribution")

plt.xticks(rotation=45)

plt.savefig("../images/payment_method.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="ReferralSource", data=df)

plt.title("Referral Source Distribution")

plt.xticks(rotation=45)

plt.savefig("../images/referral_source.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["TotalPrice"], bins=30)

plt.title("Total Price Distribution")

plt.savefig("../images/total_price_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["TotalPrice"])

plt.title("Outliers in Total Price")

plt.savefig("../images/outliers_totalprice.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="Quantity", y="TotalPrice", data=df)

plt.title("Quantity vs Total Price")

plt.savefig("../images/quantity_vs_totalprice.png")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(8,5))

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.savefig("../images/correlation_heatmap.png")

plt.show()
```

```python
top_products = df["Product"].value_counts().head(10)

print(top_products)
```

```python
plt.figure(figsize=(10,5))

sns.countplot(y="Product", data=df,
order=df["Product"].value_counts().index[:10])

plt.title("Top 10 Products")

plt.savefig("../images/top_products.png")

plt.show()
```
